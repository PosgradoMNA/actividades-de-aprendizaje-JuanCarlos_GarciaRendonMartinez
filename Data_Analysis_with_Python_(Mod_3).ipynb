# Analysis of Variance ANOVA
Es un análisis estadístico entre diferentes grupos, para saber que característica tiene más impacto en una variable. Se enfoca en analizar la varianza y busca correlaciones.

Se obtiene una prueba F: que dice la variación entre los grupos y también un P-valor que le da significancia estadística. Hay correlación fuerte si la anova nos da un valor alto de F y un valor pequeño P

# Correlación
Es una métrica estadística para saber que tan independientes son las variables. La correlación no implica la causa

Si la linea es hacia arriba la correlacion es positiva, si es hacia abajo es negativa.

# Estadísticos
Para medir la fuerza de la correlación se puede usar el método de Pearson, que nos devolverá el coeficiente y el P-valor.

Cuando los coeficientes son cercanos a 1 y -1 hay una correlación fuerte, pero cuando es 0 no. Cuando el p-valor es pequeño significa que podemos estar muy seguros con la correlación
