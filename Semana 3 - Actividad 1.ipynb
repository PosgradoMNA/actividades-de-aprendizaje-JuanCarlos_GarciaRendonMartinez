# *Parte 1: Fundamentos de bases de datos*

**Fundamentos de bases de datos y para ciencia de datos.  **
Una base de datos es el repositorio donde almacenamos datos, generalmente en forma de tablas, es decir de forma ordenada y estructurada donde cada registro contiene una lista de propiedades/datos (columnas) que lo describen. Por ejemplo podemos almacenar pagos, y por cada pago almacenar la fecha, hora, importe, folio, referencias, forma de pago, etc. Para poder acceder y recuperar esta información se usan Administradores de Base de Datos (DBMS) que nos permiten realizar peticiones y algunas operaciones a la Base de Datos y recuperar la información solicitada. Es importante para la ciencia de datos el poder manipular estos sistemas ya que el origen del a información constituye la base para generar modelos.

**Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**
Un Data Warehouse es una ampliación al concepto de Base de Datos, ya que es un lugar/almacen donde recopilamos mucha información, aquí vaciamos por ejemplo todas nuestras bases de datos de los diferentes sistemas y la idea es, además de tener todo centralizado, que podamos analizarla en conjunto, correlacionarla, generar reportes y modelos de ciencia de datos. Es decir una base de datos normal tiene el proposito fundamental de registrar información de transacciones, mientras que el análisis a profundidad lo podemos hacer a partir de un Data Warehouse
